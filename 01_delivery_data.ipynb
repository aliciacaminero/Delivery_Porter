```python
df_delivery['created_at'] = pd.to_datetime(df_delivery['created_at'])
df_delivery['actual_delivery_time'] = pd.to_datetime(df_delivery['actual_delivery_time'])
```

```python
# Verifico valores nulos por columna
print(df_delivery.isnull().sum())
```

```python
df_delivery
```

```python
df_delivery.loc[:, 'min_item_price'] = df_delivery['min_item_price'].fillna(df_delivery['min_item_price'].median())
df_delivery.loc[:, 'max_item_price'] = df_delivery['max_item_price'].fillna(df_delivery['max_item_price'].median())
```

```python
df_delivery.dropna(inplace=True)
```

***Análisis de variables numéricas***

```python
# Distribución de subtotal (precio total del pedido)

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
df_delivery['subtotal'].hist(bins=20)
plt.title('Distribución de Subtotal')
plt.xlabel('Subtotal')
plt.ylabel('Frecuencia')
plt.show()
```

```python
#Identificación de outliers:

import seaborn as sns

plt.figure(figsize=(10, 6))
sns.boxplot(x=df_delivery['subtotal'])
plt.title('Boxplot de Subtotal')
plt.show()
```

```python
Q1 = df_delivery['subtotal'].quantile(0.25)
Q3 = df_delivery['subtotal'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Eliminar outliers
df_delivery = df_delivery[(df_delivery['subtotal'] >= lower_bound) & (df_delivery['subtotal'] <= upper_bound)]
```

***Creación de nuevas variables***

```python
# Duración de la entrega
df_delivery.loc[:, 'delivery_duration'] = (
    pd.to_datetime(df_delivery['actual_delivery_time']) - pd.to_datetime(df_delivery['created_at'])
).dt.total_seconds()
```

```python
# Densidad de repartidores:
df_delivery.loc[:, 'partner_density'] = df_delivery['total_onshift_partners'] / (df_delivery['total_outstanding_orders'] + 1)
```

```python
# Día y hora del pedido
df_delivery.loc[:, 'order_day'] = pd.to_datetime(df_delivery['created_at']).dt.day_name()
df_delivery.loc[:, 'order_hour'] = pd.to_datetime(df_delivery['created_at']).dt.hour
```

***Continuo creando nuevas variables***

```python
# Duración del pedido
df_delivery.loc[:, 'delivery_duration'] = (
    pd.to_datetime(df_delivery['actual_delivery_time']) - pd.to_datetime(df_delivery['created_at'])
).dt.total_seconds()
```

```python
#Creo una columna con el periodo del día agrupando en rangos
def assign_period(hour):
    if 5 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    elif 17 <= hour < 22:
        return 'Evening'
    else:
        return 'Night'

df_delivery.loc[:, 'order_period'] = df_delivery['order_hour'].apply(assign_period)
```

```python
# Creo una columna con la carga de trabajo de los repartidores
df_delivery.loc[:, 'partner_density'] = df_delivery['total_onshift_partners'] / (df_delivery['total_outstanding_orders'] + 1)
df_delivery.loc[:, 'busy_ratio'] = df_delivery['total_busy_partners'] / (df_delivery['total_onshift_partners'] + 1)

```

```python
# Creo una columna que calcule el precio promedio por artículo
df_delivery.loc[:, 'avg_item_price'] = df_delivery['subtotal'] / df_delivery['total_items']
```

```python
# Creo otra columna en la que agrupo pedidos por el número de artículos que tienen
def order_size(items):
    if items < 3:
        return 'Small'
    elif 3 <= items < 7:
        return 'Medium'
    else:
        return 'Large'

df_delivery.loc[:, 'order_size'] = df_delivery['num_distinct_items'].apply(order_size)
```
